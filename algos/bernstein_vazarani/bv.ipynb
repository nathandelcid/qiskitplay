# Bernstein-Vazirani

We're given an oracle to a Boolean function $f: \{0,1\}^n \rightarrow \{0,1\}$, defined as the mod-2 inner product
- $s$ is hidden. **The goal is to output $s$**

$$
f(x)= s\cdot x \pmod{2}=\bigoplus_{i=0}^{n-1} s_i x_i,
$$

- $f$ takes in a bit-string $x$ and returns the **parity** of the number of positions where $s_i$ and $x_i$ are 1
    - Number of positions where $s_i$ and $x_i$ are both 1

### Think In Parity

- The oracle outputs 0 when their is **even parity** between $s$ and $x$
- The oracle outputs 1 when odd
